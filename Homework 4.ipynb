```python
def ftcs(U, V, Du, Dv, dh, nt, dt, F, k):
    
    alpha = Du*dt/(dh**2)
    beta = Dv*dt/(dh**2)

    
    for n in range (nt):
    
        Un = U.copy()
        Vn = V.copy()
        U[1:-1, 1:-1] = Un[1:-1, 1:-1]+\
            alpha*(Un[2:,1:-1]+Un[:-2, 1:-1]+Un[1:-1,2:]+Un[1:-1,:-2]-4*Un[1:-1,1:-1])-\
            Un[1:-1,1:-1]*dt*(V[1:-1,1:-1]**2)+dt*F*(1-Un[1:-1,1:-1])
        V[1:-1, 1:-1] = V[1:-1, 1:-1]+beta*(Vn[2:,1:-1]+Vn[:-2,1:-1]+Vn[1:-1,2:]+Vn[1:-1,:-2]-4*Vn[1:-1,1:-1])+\
            dt*Un[1:-1,1:-1]*(Vn[1:-1,1:-1]**2)-(F+k)*dt*Vn[1:-1,1:-1]
        #Neumann bounday conditions
        U[0, :] = U[1, :]
        U[-1,:] = U[-2,:]
        U[:, 0] = U[:, 1]
        U[:, -1] = U[:, -2]
        V[0, :] = V[1, :]
        V[-1,:] = V[-2,:]
        V[:, 0] = V[:, 1]
        V[:, -1] = V[:, -2]
    return U, V 
```

```python
#initialize
n = 192
Du, Dv, F, k = 0.00016, 0.00008, 0.035, 0.065
dh = 5./(n-1)
dx=dh
dy=dh
T = 8000
dt = 0.9*dh**2/(4*max(Du, Dv))
nt = int(T/dt)
Ui = U.copy()
Vi = V.copy()
```
